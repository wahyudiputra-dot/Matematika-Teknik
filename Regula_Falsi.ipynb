```python
import math

# 1. Definisikan Fungsi sesuai soal: f(x) = x^3 - x - 8
def f(x):
    return x**3 - 2*x - 6

# 2. Implementasi Metode Regula Falsi
def regula_falsi(a, b, e):
    step = 1
    print('\n*** IMPLEMENTASI METODE REGULA FALSI ***')

    # Cek apakah tebakan awal mengurung akar (f(a) * f(b) harus < 0)
    if f(a) * f(b) > 0.0:
        print('Tebakan awal tidak mengurung akar. Coba angka lain (misal a=2, b=3).')
        return

    condition = True
    while condition:
        # Rumus Regula Falsi untuk mencari titik potong T (x2)
        # x2 = b - (f(b) * (b - a) / (f(b) - f(a)))
        fa = f(a)
        fb = f(b)
        x2 = b - (fb * (b - a) / (fb - fa))

        fx2 = f(x2)
        print('Iterasi-%d, x2 = %0.6f dan f(x2) = %0.6f' % (step, x2, fx2))

        # Logika penggantian interval (Langkah 3 di modul)
        if fa * fx2 < 0:
            b = x2  # Akar di sub-interval bawah, maka b diganti
        else:
            a = x2  # Akar di sub-interval atas, maka a diganti

        step = step + 1
        condition = abs(fx2) > e

    print('\nAkar ditemukan pada: %0.8f' % x2)

# 3. Input Section (Bagian ini yang kamu jalankan)
# Masukkan angka: a=2, b=3, e=0.0001
try:
    input_a = float(input('First Guess a: '))
    input_b = float(input('Second Guess b: '))
    input_e = float(input('Tolerable Error (For Example 0.0001): '))

    regula_falsi(input_a, input_b, input_e)
except ValueError:
    print('Input harus berupa angka!')
```
