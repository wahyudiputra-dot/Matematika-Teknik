# 4 Variabel Gauss seidel

```python
import numpy as np

# Matriks A dan B sesuai persamaan unik di atas
a = [[12, 1, -1, 2],
     [2, 15, 2, -1],
     [1, 2, 10, 1],
     [2, -2, 1, 13]]

b = [15, -20, 35, 10]

# Pengecekan Diagonal Dominant (Persis gaya praktikum kamu)
diag = np.diag(np.abs(a))
off_diag = np.sum(np.abs(a), axis=1) - diag

if np.all(diag > off_diag):
    print('Matrix is diagonally dominant')
else:
    print('Not diagonally dominant')

# Inisialisasi variabel (Sesuai gaya Hayura)
x1 = 0
x2 = 0
x3 = 0
x4 = 0
epsilon = 0.00001
converged = False

x_old = np.array([x1, x2, x3, x4])

print('\nIteration results')
print('k,    x1,     x2,     x3,     x4 ')

for k in range(1, 50):
    # Rumus Gauss-Seidel (Update berurutan)
    x1 = (15 - 1*x2 + 1*x3 - 2*x4) / 12
    x2 = (-20 - 2*x1 - 2*x3 + 1*x4) / 15
    x3 = (35 - 1*x1 - 2*x2 - 1*x4) / 10
    x4 = (10 - 2*x1 + 2*x2 - 1*x3) / 13

    x = np.array([x1, x2, x3, x4])

    # Cek selisih (Sesuai rumus dx di praktikum kamu)
    dx = np.sqrt(np.dot(x-x_old, x-x_old))

    print("%d, %.4f, %.4f, %.4f, %.4f" % (k, x1, x2, x3, x4))

    if dx < epsilon:
        converged = True
        print('Converged!')
        break

    # Simpan nilai x terbaru ke x_old
    x_old = x.copy()

if not converged:
    print('Not converge, increase the # of iterations')
```

```python
import numpy as np

# Definisi Matriks A (Koefisien) dan Matriks B (RHS)
# Persamaan:
# 10x1 - x2 + 2x3 + 0x4 = 6
# -x1 + 11x2 - x3 + 3x4 = 25
# 2x1 - x2 + 10x3 - x1 = -11
# 0x1 + 3x2 - x3 + 8x4 = 15

a = np.array([[10, -1, 2, 0],
              [-1, 11, -1, 3],
              [2, -1, 10, -1],
              [0, 3, -1, 8]])

b = np.array([6, 25, -11, 15])

# 1. Pengecekan Diagonal Dominant (Sama seperti praktikum Hayura)
diag = np.diag(np.abs(a))
off_diag = np.sum(np.abs(a), axis=1) - diag

if np.all(diag > off_diag):
    print('Matrix is diagonally dominant - Aman untuk dilanjutkan!')
else:
    print('Not diagonally dominant - Mungkin tidak konvergen')

# 2. Inisialisasi Variabel
x1 = 0; x2 = 0; x3 = 0; x4 = 0
epsilon = 0.00001
converged = False
x_old = np.array([x1, x2, x3, x4])

print('\nIteration results (Gauss-Seidel 4x4)')
print(' k |    x1    |    x2    |    x3    |    x4   ')
print('-' * 45)

# 3. Loop Iterasi Gauss-Seidel
for k in range(1, 50):
    # Update masing-masing variabel secara berurutan
    # Menggunakan nilai terbaru yang sudah dihitung (Ciri khas Gauss-Seidel)
    x1 = (6 + x2 - 2*x3) / 10
    x2 = (25 + x1 + x3 - 3*x4) / 11
    x3 = (-11 - 2*x1 + x2 + x4) / 10
    x4 = (15 - 3*x2 + x3) / 8

    x = np.array([x1, x2, x3, x4])

    # Hitung selisih (Error) menggunakan Euclidean Distance
    dx = np.sqrt(np.dot(x - x_old, x - x_old))

    print(f"{k:2d} | {x1:8.4f} | {x2:8.4f} | {x3:8.4f} | {x4:8.4f}")

    if dx < epsilon:
        converged = True
        print('-' * 45)
        print(f'Konvergen pada iterasi ke-{k}!')
        break

    x_old = x.copy()

if not converged:
    print('Tidak konvergen setelah 50 iterasi.')
```
